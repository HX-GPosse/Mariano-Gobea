Para rellenar estos valores, tenemos dos opciones:

Y por ultimo tenemos la columna de duration, esta columna contiene los datos de la duracion de la pelicula o serie correspondiente

Vemos que, como la columna de rating, tiene un porcentaje muy bajo de valores vacios 0.035 %

```python
df[df['duration'].isnull() == True]
# Son solo 3 registros los vacios
```

Para trabajar estos tres registros vacios, vamos a tener varias opciones:

* Rellenar con un valor especifico, por ejemplo : "Unknown" Para hacer referencia de que esos registros no tienen valor en duracion

* Rellenar con alguna medida central, moda media mediana,  de los datos de la columna

* Rellenar basado en el tipo de show.

Vamos a ir por la tercera opcion, pero vale recalcar que si tuvieramos acceso a datos externos podriamos complementar con estos valores para tener precisión con los datos


```python
# Lo primero que vamos a realizar es separar en dos DataFrames los diferentes tipos de show que tenemos
df['type'].drop_duplicates()

# Solo tenemos dos tipos de show, Movie y TV Show, asi que solo debemos crear 2 DataFrames nuevos
```

```python
movies = df[df['type'] == 'Movie']

series = df[df['type'] == 'TV Show']

# Una vez separados en los DF lo que vamos a hacer es combinar con la tercera opcion mencionada anteriormente, Rellenar con el valor modal. 
# Pero primero debemos obtener los valores de la moda para las Peliculas y la moda para las Series

# Calculamos los valores modales y los almacenamos en variables separadas 
movies_moda = movies['duration'].mode()[0]
series_moda = series['duration'].mode()[0]
```

```python
# Ahora solo nos queda rellenar los valores en las filas que contenian estos NaN. Para ello vamos a crear una funcion que nos va a ayudar con este proceso

def fill_duration(fila, moda_movies, moda_series):
    """
    Esta funcion recibe tres parametros:
    fila: Será la fila que estamos por modificar
    moda_movies: Será el valor modal de los registros de tipo Movies
    moda_series: Será el valor modal para los registros de tipo TV Shows
    """
    if pd.isna(fila['duration']):
        if fila['type'] == 'Movie':
            return moda_movies
        elif fila['type'] == 'TV Show':
            return moda_series
    return fila['duration']


# Y ahora utilizando el metodo apply de los DataFrame vamos a poder aplicar esta funcion para cada uno de los valores de Duration
df['duration'] = df.apply(lambda row: fill_duration(row, movies_moda, series_moda), axis=1)
```

```python
# Chequeamos que no haya quedado ninguna fila con valores vacios
df[df['duration'].isnull() == True]
```

Esto no termina acá, vamos a realizar algunas visualizaciones para ver como esta la distribucion de nuestros datos y hacer un buen analisis previo


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style()
```

```python
colors = sns.color_palette("cubehelix", 2)
df["type"].value_counts().plot.pie(autopct='%1.2f%%',explode=[0,0.08], shadow = True, colors=colors)
plt.show()
```

Vamos a arrancar con este gráfico que nos va a enseñar la distribucion de los datos en relación al Tipo de Show 

Se observa a simple vista que las Películas predominan a la plataforma de Netflix, superando a las Series con un 69,6 %

```python
df['year_added'] = df['date_added'].dt.year
# Este codigo nos va a crear una nueva columna con unicamente los valores de los años de cuando se subieron los titulos a Netflix
# Podemos omitirlo para la data final pero nos va a servir para el siguiente grafico
```

```python
df['year_added']
```

```python
# Agrupamos los datos por año de carga y tipo de contenido, y contamos el número de títulos en cada grupo
titulos = df.groupby(['year_added','type']).size().unstack(fill_value=0)
# Reiniciamos el índice del DataFrame para que 'year_added' vuelva a ser una columna
titulos = titulos.reset_index()
# Transformamos el DataFrame para tener una estructura más adecuada para el gráfico, con columnas 'year_added', 'Type' y 'Conteo'
titulos = titulos.melt(id_vars='year_added', var_name='Type', value_name='Conteo')


# Configuramos el lienzo sobre el cual vamos a hacer el gráfico
plt.figure(figsize=(24,6))
plt.title('Titulos añadidos a Netflix por año y por tipo de contenido')

# Creamos el gráfico de barras
ax = sns.barplot(x=titulos['year_added'], y=titulos['Conteo'], hue=titulos['Type'], palette='cubehelix')

# Añadimos etiqueta al eje x
# Rotamos las etiquetas del eje x para una mejor visualización
plt.xlabel('Año de carga en Netflix', fontsize=16)
plt.xticks(rotation=45)

# Añadimos etiqueta al eje y
plt.ylabel('Conteo', fontsize=16)

# Añadimos etiquetas a las barras para mostrar el valor de cada barra
for lab in ax.containers:
    ax.bar_label(lab)
```

Este gráfico, como dice el título, nos permite ver la cantidad de Peliculas y Series que fueron añadidas a la plataforma de Netflix a traves de los diferentes años.

Esta compuesto por barras y se puede diferencias los tipos de títulos, Movie y TV Show, por los colores que están especificados arriba a la derecha.

Podemos ver que el año en donde más títulos se añadieron fue en 2019 con un total de 2016 titulos, 1424 Peliculas y 592 Series

```python
# Filtramos el DataFrame para obtener únicamente las filas correspondientes a películas
movies = df[df['type'] == 'Movie']
# Eliminamos la cadena ' min' de la columna 'duration' y convertimos los valores a enteros
movies['duration'] = movies['duration'].str.replace(' min','').astype(int)
# Calculamos la duración promedio de las películas y redondeamos a dos decimales
duracion_promedio = round(np.mean(movies['duration']), 2)


# Configuramos el lienzo para el gráfico
plt.figure(figsize=(12,6))

plt.title('Duración de las películas (minutos)', fontsize=24)

# Creamos un histograma de la duración de las películas con una línea de densidad
sns.histplot(movies['duration'], kde=True, color='midnightblue')

plt.axvline(x=duracion_promedio, color='red', linestyle='--', label= f'Duración Promedio {duracion_promedio} min')

plt.xlabel('Duración de la pelicula (min)', fontsize=16)

plt.ylabel('Cantidad de películas', fontsize=16)

plt.legend()

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()

plt.show()
```

Este gráfico es un histograma que nos muestra la cantidad de películas segun la Duración en minutos de las mismas.

Podemos ver tambien la Duración promedio que es de 99.57 minutos, esta señalizado con una linea punteada en el gráfico.

La mayoria de las peliculas se encuentra cerca de estos valores, vemos que ya son muy pocas las peliculas que duran mas de 120 minutos

Tambien cuenta con una linea azul que nos muestra la densidad en el histograma y nos muestra cómo los valores de la duración de las películas están distribuidos de manera suave a lo largo de su rango.
A diferencia de las barras del histograma, que muestran la frecuencia de los datos en intervalos discretos, la línea de densidad proporciona una visión continua de la distribución de los datos

```python
series = df[df['type'] == 'TV Show']
series['duration'] = series['duration'].str.extract('(\d+)').astype(int)
duracion_promedio = round(np.mean(series['duration']), 2)

plt.figure(figsize=(12,6))
plt.title('Duración series en Netflix')
sns.histplot(series['duration'], bins=20, color='midnightblue')
plt.axvline(x=duracion_promedio, color='RED', linestyle='--', label=f'Duración promedio {duracion_promedio} temporadas')
plt.xlabel('Duración en temporadas')
plt.ylabel('Cantidad Series')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
```

Como antes habiamos hecho el gráfico de duración de las películas, es importante tambien comparar las series o TV Show dentro de nuestro Dataset

Podemos ver que la mayoria de las Series no tienen mas de 4 temporadas, siendo un bajo porcentaje las que tienen 6 temporadas o más de duración

Tambien, podemos visualizar que la duración promedio es de 1.76 temporadas


```python
# Convertir la columna 'date_added' a tipo datetime
df['date_added'] = pd.to_datetime(df['date_added'], format='%Y-%m-%d', errors='coerce')

# Extraer el año de la columna 'date_added'
df['year_added'] = df['date_added'].dt.year

# Contar la cantidad de títulos añadidos por año
titles_per_year = df['year_added'].value_counts().sort_index()

# Configurar el lienzo para el gráfico
plt.figure(figsize=(14,8))
plt.title('Cantidad de Títulos Añadidos a Netflix por Año', fontsize=16)

# Crear el gráfico de líneas para mostrar la tendencia
sns.lineplot(x=titles_per_year.index, y=titles_per_year.values, marker='o', color='midnightblue')

# Añadir anotaciones en cada punto
for x, y in zip(titles_per_year.index, titles_per_year.values):
    plt.annotate(f'{y}', xy=(x, y), xytext=(0, 5), textcoords='offset points', ha='center', fontsize=12)

plt.xlabel('Año', fontsize=14)
plt.ylabel('Cantidad de Títulos', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
